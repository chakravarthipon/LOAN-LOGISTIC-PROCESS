```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import roc_auc_score,accuracy_score
from sklearn.model_selection import train_test_split
```

```python
RFClass = RandomForestClassifier() # it was used to hypertune with cross parameters
NNClass = MLPClassifier()
LRClass = LogisticRegression()
```

```python
loan = pd.read_csv('home_sale.csv')
loan
```

```python
loan.describe()
```

```python
loan_feat =loan.iloc[:,0:2]
loan_targ =loan.iloc[:,2]
loan_feat
```

```python
loan_targ.head()
```

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
```

```python
loan_targLE = LabelEncoder().fit_transform(loan['Loan'])
```

```python
loan_targLE
```

```python
loanTrainx,loanTestx,loanTrainy,loanTesty =train_test_split(loan_feat,loan_targLE,test_size =0.4)


#fit trainning data.Train models on training data
RFClass.fit(loanTrainx,loanTrainy)
NNClass.fit(loanTrainx,loanTrainy)
LRClass.fit(loanTrainx,loanTrainy)


#Prediction out on test dataset
RFpredLoan = RFClass.predict(loanTestx)
NNpredLoan = NNClass.predict(loanTestx)
LRpredLoan = LRClass.predict(loanTestx)


#Error metrics using roc_auc score
print('RF Classifier roc_auc score:',roc_auc_score(loanTesty,RFpredLoan))
print('NN Classifier roc_auc score:',roc_auc_score(loanTesty,NNpredLoan))
print('LR Classifier roc_auc score:',roc_auc_score(loanTesty,LRpredLoan))


#Error metrics using accuracy score
print('RF Classifier accuracy score:',accuracy_score(loanTesty,RFpredLoan))
print('NN Classifier accuracy score:',accuracy_score(loanTesty,NNpredLoan))
print('LR Classifier accuracy score:',accuracy_score(loanTesty,LRpredLoan))


```

```python
# Previous code...

# Error metrics using accuracy score
rf_accuracy = accuracy_score(loanTesty, RFpredLoan)
nn_accuracy = accuracy_score(loanTesty, NNpredLoan)
lr_accuracy = accuracy_score(loanTesty, LRpredLoan)

print('RF Classifier accuracy score: {:.2f}%'.format(rf_accuracy * 100))
print('NN Classifier accuracy score: {:.2f}%'.format(nn_accuracy * 100))
print('LR Classifier accuracy score: {:.2f}%'.format(lr_accuracy * 100))
```
